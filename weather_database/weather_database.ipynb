```python
import numpy as np
import pandas as pd
from citipy import citipy
import requests
import sys, os
```

```python
# stack exchange answer from user aruisdante in 2014:
# https://stackoverflow.com/questions/24868733/how-to-access-a-module-from-outside-your-file-folder-in-python
sys.path.append(os.path.join('..'))
from config import weather_api_key
```

```python
rand_lats = [np.random.uniform(-90, 90) for i in range(2000)]
rand_longs = [np.random.uniform(-180, 180) for i in range(2000)]
```

```python
rand_cities = [citipy.nearest_city(rand_lats[e], rand_longs[e]).city_name for e in range(2000)]
```

```python
weather_url = 'http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=' + weather_api_key + '&q='
```

```python
city_weather = requests.get(weather_url + rand_cities[4]).json()
```

```python
city_weather
```

```python
weather_data_list = []

for city in rand_cities:
    
    # formatting the city name to avoid 404s
    city.replace(' ', '')
    
    try:
        # try to call api for that city
        city_weather = requests.get(weather_url + city).json()
    
        # create dict for each city
        weather_dict = {
            'city': city,
            'lon': city_weather['coord']['lon'],
            'lat': city_weather['coord']['lat'],
            'max_temp': city_weather['main']['temp_max'],
            'humidity': city_weather['main']['humidity'],
            'cloudiness': city_weather['clouds']['all'],
            'wind_speed': city_weather['wind']['speed'],
            'weather': city_weather['weather'][0]['description']}
    
        # append the dict for each city to the list of city weather
        weather_data_list.append(weather_dict)
    except:
        print(f'{rand_cities.index(city)} -- no response for {city}')
```

```python
pd.DataFrame(weather_data_list).to_csv('weatherpy_database.csv', index_label='city_id')
```
